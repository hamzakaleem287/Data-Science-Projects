### Explaining the Dataset Terms

##### cp: 
This stands for chest pain type. It could be typical angina, atypical angina, non-anginal pain, or asymptomatic.
##### trestbps: 
This refers to resting blood pressure, measured in millimeters of mercury (mm Hg).
##### chol: 
This represents serum cholesterol levels in mg/dl (milligrams per deciliter).
##### fbs:
This stands for fasting blood sugar. It indicates the level of glucose in the blood after fasting for at least 8 hours. It's measured in mg/dl.
##### restecg:
This refers to resting electrocardiographic results. It describes the results of an electrocardiogram (ECG or EKG) while the person is at rest.
##### thalach: 
This denotes the maximum heart rate achieved during the exercise test. It's measured in beats per minute (bpm).
##### exang: 
This stands for exercise-induced angina. It indicates whether angina (chest pain) was induced by exercise. It's typically a binary variable (yes/no or 1/0).
##### oldpeak: 
This represents the ST depression induced by exercise relative to rest. It's measured in mm (millimeters) on an ECG reading.
##### slope: 
This describes the slope of the peak exercise ST segment. It can be upsloping, flat, or downsloping.
##### ca: 
This denotes the number of major vessels (0-3) colored by fluoroscopy. It represents the number of major blood vessels supplying the heart with blood that are visible during fluoroscopy.
##### thal: 
This refers to thallium stress test. It categorizes the results of the thallium stress test as normal, fixed defect (no change in the results during exercise), or reversible defect (a partial or complete reversal of the defect during exercise).
##### target: 
This is the target variable. In the context of heart disease prediction, it typically indicates whether or not a patient has heart disease.
