### Explaining the Terms Used

##### cogs:
"COGS" stands for Cost of Goods Sold.It includes the cost of materials, labor, and other expenses directly associated with producing the goods sold.
##### gross margin percentage:
A higher gross margin percentage indicates better profitability, as it means that the company is earning more from each sale after accounting for the direct costs.
##### gross income
Gross income provides insight into the overall profitability of sales operations.

### Display the Daily sales by the Day of the WeeK

```python
data.columns
```

```python
dw_mapping={0:"Mon",1:"Tue",2:"Wed",3:"Thur",4:"Fri",5:"Sat",6:"Sun"}
data["Day_of_Week"]=data["Date"].dt.dayofweek.map(dw_mapping)
```

```python
data["Day_of_Week"].value_counts().plot(kind="bar")
```

### What will be the highest month for sales

```python
month_mapping={1:"Jan",2:"Feb",3:"Mar",4:"Apr"}
data["Month"]=data["Date"].dt.month.map(month_mapping)
```

```python
data["Month"].value_counts().plot(kind="bar")
plt.show()
```
